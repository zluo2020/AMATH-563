```python
from scipy.io import loadmat
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
data = loadmat('kuramoto_sivishinky_4.mat')
```

```python
uu = data['uu']
tt = data['tt']
x = data['x']
```

```python
uu.shapeK-S
```

```python
plt.pcolormesh(uu.T,cmap='hot',shading='flat')
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(tt, x, uu, cmap=cm.coolwarm)
```

```python
x_train = uu.T[:-1, :]
y_train = uu.T[1:, :]
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(1024, activation='linear'))
model.compile(loss='mse', optimizer='adam', validation_split=0.2)
nn = model.fit(x_train, y_train, epochs=1000)
```

```python
y_pred = np.zeros(y_train.shape)
uu0 = x_train[0]

for i in range(x_train.shape[0]):
    
    pred = model.predict(uu0.reshape((1,1024)))
    y_pred[i,:] = pred
    uu0 = pred
    
```

```python
plt.subplot(1, 2, 1)
plt.pcolormesh(y_train,cmap='hot',shading='flat')
plt.title('Original')
plt.subplot(1, 2, 2)
plt.pcolormesh(y_pred,cmap='hot',shading='flat')
plt.title('NN')
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(tt[0][:-1], x, y_pred.T, cmap=cm.coolwarm)
ax.set_title('NN')
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(tt[0][:-1], x, y_train.T, cmap=cm.coolwarm)
ax.set_title('Original')
```
